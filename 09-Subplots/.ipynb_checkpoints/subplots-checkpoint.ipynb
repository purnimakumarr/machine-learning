### AIM
#### Generate different subplots from a given plot and color plot data.

### CODE & OUTPUT

```python
import numpy as np
from matplotlib import pyplot as plt

# SAMPLE DATA
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
```

```python
# A SINGLE SUBPLOT
fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("A single plot")
```

```python
# subplots in stacked in vertical direction
fig, (axs1, axs2) = plt.subplots(2)
axs1.plot(x, y, "tab:green")
axs1.set(xlabel="x-label", ylabel="y_label")
axs2.plot(x, -y, "tab:orange")
```

```python
# subplots in stacked in horizontal direction
fig, (axs1, axs2) = plt.subplots(1, 2)
axs1.plot(x, y, "tab:red")
axs1.set(xlabel="x-label", ylabel="y_label")
axs2.plot(x, -y)
```

```python
# Stacking subplots in two directions
fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Axis [0, 0]')
axs[0, 1].plot(x, y, 'tab:orange')
axs[0, 1].set_title('Axis [0, 1]')
axs[1, 0].plot(x, -y, 'tab:green')
axs[1, 0].set_title('Axis [1, 0]')
axs[1, 1].plot(x, -y, 'tab:red')
axs[1, 1].set_title('Axis [1, 1]')

for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```
